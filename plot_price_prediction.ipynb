```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

data = {
    "Area": [500, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000],
    "Price": [100000, 150000, 200000, 250000, 300000, 350000, 400000, 450000, 500000, 550000]
}
df = pd.DataFrame(data)


X = df[["Area"]]     # Make sure it's 2D
y = df["Price"]      # 1D


model = LinearRegression()
model.fit(X, y)


y_pred = model.predict(X)


r2 = r2_score(y, y_pred)
print("R² Score:", r2)
```

```python
plt.scatter(X, y, color='red', label='Actual Prices')
plt.plot(X, y_pred, color='yellow', linewidth=2, label='Predicted Line')
plt.xlabel("Area (sq ft)")
plt.ylabel("Price (₹)")
plt.title("Linear Regression: Area vs. Price")
```
